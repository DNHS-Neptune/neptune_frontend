# Binary Calculator

```python

# Get inputs from user
b1 = input("Enter the first binary number (e.g., 1010): ")
op = input("Enter an operator (+, -, *, //): ")
b2 = input("Enter the second binary number (e.g., 0011): ")

# Convert binary strings to integers
try:
    d1 = int(b1, 2)
    d2 = int(b2, 2)
except ValueError:
    print("❌ Invalid binary input.")
    result = None
else:
    # Perform operation
    if op == "+":
        result = d1 + d2
    elif op == "-":
        result = d1 - d2
    elif op == "*":
        result = d1 * d2
    elif op == "//":
        if d2 == 0:
            print("❌ Cannot divide by zero.")
            result = None
        else:
            result = d1 // d2
    else:
        print("❌ Invalid operator.")
        result = None

# Print the result
if result is not None:
    print(f"{b1} {op} {b2} = {bin(result)[2:]} (binary)")
```

# Decimal to Binary

```python

# Get user input
try:
    decimal = int(input("Enter a decimal number (e.g., 13): "))
    binary = bin(decimal)[2:]  # Convert to binary, strip the '0b'
    print(f"{decimal} in binary is {binary}")
except ValueError:
    print("❌ Please enter a valid integer.")
```
