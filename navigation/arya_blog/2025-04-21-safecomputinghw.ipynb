# Popcorn Hack #2: Dice Roll

```python
import random

def biased_color():
    """Generates a biased random color.
    Red: 50%, Blue: 30%, Others: 20%
    """
    colors = ["Red", "Blue", "Green", "Yellow", "Purple"]
    probabilities = [0.5, 0.3, 0.07, 0.07, 0.06] 

    return random.choices(colors, probabilities)[0]

for _ in range(10):
    print("Color:", biased_color())
```

# Homework Hack

```python
import random

def coin_flip():
    """Simulates a single coin flip."""
    return random.choice(["heads", "tails"])

def coin_flip_game():
    player1_heads = 0
    player2_heads = 0
    rounds = 0

    while player1_heads < 3 and player2_heads < 3:
        rounds += 1
        p1 = coin_flip()
        p2 = coin_flip()

        if p1 == "heads":
            player1_heads += 1
        if p2 == "heads":
            player2_heads += 1

        print(f"Round {rounds}: Player 1 - {p1} ({player1_heads} heads), Player 2 - {p2} ({player2_heads} heads)")

    winner = "Player 1" if player1_heads == 3 else "Player 2"
    print(f"\n🏆 {winner} wins after {rounds} rounds!")

coin_flip_game()
```
