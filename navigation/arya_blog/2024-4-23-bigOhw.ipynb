# Time Complexity Examples HW

## O(1) - Constant Time

```python
# O(1) - Constant Time
def get_first_element(arr):
    return arr[0]

arr = [10, 20, 30, 40, 50]
print("First element:", get_first_element(arr))
```

## O(log n) - Logarithmic Time

```python
# O(log n) - Logarithmic Time
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

arr = [1, 3, 5, 7, 9, 11, 13, 15]
print("Index of 7:", binary_search(arr, 7))
```

## O(n) - Linear Time

```python
# O(n) - Linear Time
def linear_search(arr, target):
    for i, val in enumerate(arr):
        if val == target:
            return i
    return -1

arr = list(range(100))
print("Index of 45:", linear_search(arr, 45))
```

## O(n²) - Quadratic Time

```python
# O(n^2) - Quadratic Time
def print_all_pairs(arr):
    for i in arr:
        for j in arr:
            print(f"({i}, {j})")

arr = [1, 2, 3]
print_all_pairs(arr)
```
