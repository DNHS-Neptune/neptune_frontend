## 1. Undecidable Problem Research
**Problem:** *Post Correspondence Problem (PCP)*

The PCP is an undecidable decision problem introduced by Emil Post in 1946. The problem involves matching sequences of strings from two lists to determine whether a sequence of indices exists such that the concatenation of strings from the first list equals the concatenation from the second list.

### Example:
List A = ["a", "ab"]  
List B = ["aa", "b"]

Check if there is a sequence of indices (with repetition) so that the strings from both lists match when concatenated in that sequence.

This problem has been proven undecidable for general inputs.

```python
# 2. Nearest Neighbor Algorithm Implementation for TSP

def nearest_neighbor_tsp(distances, start=0):
    n = len(distances)
    unvisited = set(range(n))
    unvisited.remove(start)
    tour = [start]
    current = start
    total_distance = 0

    while unvisited:
        nearest = min(unvisited, key=lambda city: distances[current][city])
        total_distance += distances[current][nearest]
        current = nearest
        tour.append(current)
        unvisited.remove(current)

    total_distance += distances[current][start]
    tour.append(start)

    return tour, total_distance

# Distance matrix for cities A-E
distances = [
    [0, 12, 19, 8, 15],
    [12, 0, 10, 18, 5],
    [19, 10, 0, 6, 14],
    [8, 18, 6, 0, 9],
    [15, 5, 14, 9, 0]
]

# Start from city A (index 0)
tour, distance = nearest_neighbor_tsp(distances)
tour, distance
```

## 3. Real-World Example Using Heuristics
**Example:** Google Maps and GPS Navigation

GPS systems like Google Maps use heuristic-based algorithms (like A*) to compute driving directions.

### Why Not Exact Solutions?
- Road networks can be extremely large (millions of intersections).
- Traffic data is dynamic and changes constantly.
- Computing the shortest path exactly for all possibilities would be computationally infeasible in real time.

**Heuristics offer fast, good-enough paths** that consider distance, time, and traffic.
