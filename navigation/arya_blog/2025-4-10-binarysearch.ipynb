# Binary Search on a Rotated Array

```python
def search_rotated_array(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        
        # Check if the target is found
        if arr[mid] == target:
            return mid

        # Left half is sorted
        if arr[left] <= arr[mid]:
            if arr[left] <= target < arr[mid]:  # Target is in the left half
                right = mid - 1
            else:  # Target is in the right half
                left = mid + 1
        # Right half is sorted
        else:
            if arr[mid] < target <= arr[right]:  # Target is in the right half
                left = mid + 1
            else:  # Target is in the left half
                right = mid - 1

    return -1
arr = [4, 5, 6, 7, 0, 1, 2]
target = 1
print(search_rotated_array(arr, target))  # Output: 5

```

# First and Last Occurance of an Element

```python
def find_first_last(arr, target):
    # Function to find the first occurrence
    def find_first():
        left, right = 0, len(arr) - 1
        first = -1
        while left <= right:
            mid = (left + right) // 2
            if arr[mid] == target:
                first = mid
                right = mid - 1
            elif arr[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
        return first

    # Function to find the last occurrence
    def find_last():
        left, right = 0, len(arr) - 1
        last = -1
        while left <= right:
            mid = (left + right) // 2
            if arr[mid] == target:
                last = mid
                left = mid + 1
            elif arr[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
        return last

    first = find_first()
    last = find_last()

    return (first, last) if first != -1 else -1

arr = [1, 2, 2, 2, 3, 4, 5]
target = 2
print(find_first_last(arr, target))  # Output: (1, 3)
```

# Search for Smallest Element

```python
def smallest_ge(arr, target):
    left, right = 0, len(arr) - 1
    result = -1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] >= target:
            result = arr[mid]
            right = mid - 1
        else:
            left = mid + 1

    return result
arr = [1, 3, 5, 7, 9, 11]
target = 8
print(smallest_ge(arr, target)) 

```
